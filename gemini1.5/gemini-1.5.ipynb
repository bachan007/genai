```python
import google.generativeai as genai

import os
from dotenv import load_dotenv
load_dotenv()

genai.configure(api_key=os.getenv('GEMINI-API-KEY'))
```

```python
type(genai.list_models())
```

```python
for model in genai.list_models():
    # print(model.supported_generation_methods)
    if 'generateContent' in model.supported_generation_methods:
        print(model.name)
```

```python
model = genai.GenerativeModel('gemini-1.5-pro-latest')
```

```python
response = model.generate_content('why google was renamed as Alphabet?')
```

```python
response.text
```

```python
response = model.generate_content('What is the history of Large Language Models? I need a detailed explanation.',stream=True)
```

```python
for chunk in response:
    print(chunk.text)
    print('\n')
    
```

```python
with open('output.txt','w') as f:
    f.write(response.text)
f.close()
```
